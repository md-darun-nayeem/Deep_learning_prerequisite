```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

# Distribution plot :
- figure lavel: displot
- axes level : hsitplot, kdeolot, ruhplot

```python
# plotting univariate histogram
tips

```

```python
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20,hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20,hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',bins=20,hue='sex')
```

```python
# faceting using col and row --> not work in histplot function
sns.displot(data=tips,x='total_bill',kind='hist',col='sex',element='step')
```

```python
sns.kdeplot(data = tips, x='total_bill')
```

```python
sns.displot(data = tips, x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
sns.histplot(data=tips,x='total_bill',y='tip')
```

```python
sns.kdeplot(data=tips,x='total_bill',y='tip')
```
