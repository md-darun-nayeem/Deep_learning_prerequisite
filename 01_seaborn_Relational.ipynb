Two types of functions in seaborn:
- Figure level
- Axis level

Types of graphs in seaborn:
- scatter plot
- line plot
- distribution plot --> Histogram | kde_plot | rug_plot
- catagorical plot --> Barplot | boxplot | violin | count plot | swamplot
- matrix plot --> heatmap | claster map
- regression plot
- multiplots --> joint plot | pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
# scatter plot : axis level function
sns.scatterplot(data = tips, x ='total_bill', y ='tip',hue='sex',style='time',size="size")
```

```python
# relplot : figure level function --> square
sns.relplot(data = tips, x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'Bangladesh']
temp_df
```

```python
# axis level function
sns.lineplot(data = temp_df, x= 'year', y = 'lifeExp')
```

```python
# relplot
sns.relplot(data = temp_df, x = 'year', y= 'lifeExp', kind = 'line')
```

```python
temp_data = gap[gap['country'].isin(['Bangladesh','Pakistan','India'])]
temp_data
```

```python
sns.relplot(kind = 'line', data=temp_data, x='year', y='lifeExp', hue='country')
```

```python
temp_data = gap[gap['country'].isin(['Bangladesh','Brazil','Germany'])]
temp_data
```

```python
sns.lineplot( data=temp_data, x='year', y='lifeExp', hue='country',style='continent')
```

```python
# facet plot
sns.relplot(data= tips, x='total_bill',y='tip',kind='scatter',col='sex',row='smoker')
```

```python
# facet plot , kind = line
# It will work with relplot, not work with scatter plot and line plot
sns.relplot(data= tips, x='total_bill',y='tip',kind='line',col='sex',row='smoker')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```
