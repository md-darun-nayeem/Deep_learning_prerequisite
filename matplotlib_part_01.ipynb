```python
# 2D plots

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
price = [5000,4000,8000,7000,4525,7895]
year = [2020,2021,2022,2023,2024,2025]
plt.plot(year,price)
```

```python
batsman = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python

plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed',marker='D',label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#12E3EA',linestyle='dotted',marker='D',label='rohit')
plt.title('Virat vs Rohit')
plt.xlabel('Season')
plt.ylabel('Runs')
plt.legend()
```
