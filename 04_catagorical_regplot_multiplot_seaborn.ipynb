```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

Catogorical plot:
- catagorical scatter plot -->stripplot, swarmplot
- catagorical distribution plot --> box plot, violin plot
- catagorical estimate plot --> fot center tendency
        : Bar plot, point plot, count plot
- figure level function ---> cat plot

```python
sns.scatterplot(data=tips, x='day', y='total_bill')
```

```python
sns.stripplot(data= tips,x='day', y='total_bill',jitter=False)  # Here jitter is additional noise
```

```python
sns.catplot(data=tips, x='day', y='total_bill',kind='strip',jitter=False, hue='sex')
```

```python
# swarm plot : helpful for smaller data set
sns.catplot(data=tips, x='day' ,y='total_bill',kind='swarm')
```

```python
sns.swarmplot(data= tips , x= 'day', y='total_bill')
```

```python
# Box plot

sns.boxplot(data=tips, x='day',y='total_bill')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='box', hue='sex')
```

```python
sns.boxplot(data=tips, y='total_bill')
```

```python
# violinplot = (Boxplot + KDEplot)
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
# catagorical estimate plot
import numpy as np

sns.barplot(data=tips,x='sex',y='total_bill',hue='smoker',estimator=np.min)
```

```python
# point plot : helps to find out difference between two catagorie
sns.pointplot(data=tips,x='sex',y='total_bill')
```

```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
# regplot : linear regression plot : same as lmplot
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
# residplot : find out the error of a regression model
sns.residplot(data=tips,x='total_bill',y='tip')
```

```python
# figure level -> relplot,displot,catplot,lmplot
```

```python
# Fasit grid

g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

```python
# pair plot
sns.pairplot(tips,hue='sex')
```

```python
# Jointplot
sns.jointplot(data=tips,x='total_bill',y='tip')
```
